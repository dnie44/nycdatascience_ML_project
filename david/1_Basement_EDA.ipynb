```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing the data
housing = pd.read_csv('All_Data.csv')
```

## Features
___

### BsmtQual: Evaluates the height of the basement
       
       Ex   Excellent (100+ inches)	
       Gd   Good (90-99 inches)
       TA   Typical (80-89 inches)
       Fa   Fair (70-79 inches)
       Po   Poor (<70 inches
       NA   No Basement
       
### BsmtCond: Evaluates the general condition of the basement

       Ex   Excellent
       Gd   Good
       TA   Typical - slight dampness allowed
       Fa   Fair - dampness or some cracking or settling
       Po   Poor - Severe cracking, settling, or wetness
       NA   No Basement
       
### BsmtExposure: Refers to walkout or garden level walls

       Gd   Good Exposure
       Av   Average Exposure (split levels or foyers typically score average or above)	
       Mn   Mimimum Exposure
       No   No Exposure
       NA   No Basement
       
### BsmtFinType1: Rating of basement finished area

       GLQ  Good Living Quarters
       ALQ  Average Living Quarters
       BLQ  Below Average Living Quarters	
       Rec  Average Rec Room
       LwQ  Low Quality
       Unf  Unfinshed
       NA   No Basement
       
### BsmtFinType2: Rating of basement finished area (if multiple types)

       GLQ  Good Living Quarters
       ALQ  Average Living Quarters
       BLQ  Below Average Living Quarters	
       Rec  Average Rec Room
       LwQ  Low Quality
       Unf  Unfinshed
       NA   No Basement

```python
basement_df = pd.DataFrame(housing[['PID',
                      'SalePrice',
                      'BsmtQual',
                      'BsmtCond',
                      'BsmtExposure',
                      'BsmtFinType1',
                      'BsmtFinType2']])
basement_df = basement_df.fillna('NoBasement')
```

```python
basement_height = pd.DataFrame(basement_df[['SalePrice','BsmtQual']])
fig = px.box(basement_height, 
             x = 'BsmtQual', 
             y = 'SalePrice',
             color = 'BsmtQual',
            title = 'Home Sale Price by Basement Height')
fig.show()
```

```python
basement_quality = pd.DataFrame(basement_df[['SalePrice','BsmtCond']])
fig2 = px.box(basement_quality, 
             x = 'BsmtCond', 
             y = 'SalePrice',
             color = 'BsmtCond',
            title = 'Home Sale Price by Basement Condition')
fig2.show()
```

```python
basement_access = pd.DataFrame(basement_df[['SalePrice','BsmtExposure']])
fig3 = px.box(basement_access, 
             x = 'BsmtExposure', 
             y = 'SalePrice',
             color = 'BsmtExposure',
            title = 'Home Sale Price by Basement Access')
fig3.show()
```

```python
basement_type1 = pd.DataFrame(basement_df[['SalePrice','BsmtFinType1']])
fig4 = px.box(basement_type1, 
             x = 'BsmtFinType1', 
             y = 'SalePrice',
             color = 'BsmtFinType1',
            title = 'Home Sale Price by Basement Type 1')
fig4.show()
```

```python
basement_type2 = pd.DataFrame(basement_df[['SalePrice','BsmtFinType2']])
fig5 = px.box(basement_type2, 
             x = 'BsmtFinType2', 
             y = 'SalePrice',
             color = 'BsmtFinType2',
            title = 'Home Sale Price by Basement Type 2')
fig5.show()
```
