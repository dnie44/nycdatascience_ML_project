```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import tree
import plotly.express as px
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
ah = pd.read_csv('Ames_Housing_Price_Data.csv', index_col=0,low_memory=False)
```

<h3>Fireplaces EDA</h3>

```python
fig = px.scatter(ah,x='GrLivArea', y='SalePrice',color='Fireplaces')
fig.show()
```

**Seems to be many strong linear relationships between number of fireplaces and home value** <br> <br>
**Lets look at houses that are between 1250 and 2250 in above ground square feet**

```python
zoom = ah[((ah[['GrLivArea']] > 1250) & (ah[['GrLivArea']] <= 2250)).any(axis=1)]
```

```python
fig = px.scatter(zoom,x='GrLivArea', y='SalePrice',color='Fireplaces',trendline="ols")
fig.show()
```
